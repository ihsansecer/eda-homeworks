## Question 5

```python
neverMarried = resp[resp["fmarno"] == 0]
married = resp[resp["fmarno"] > 0]
```

```python
nm_hist = thinkstats2.Hist(neverMarried["totincr"])
m_hist = thinkstats2.Hist(married["totincr"])

w = 0.4
thinkplot.PrePlot(2)
thinkplot.Hist(nm_hist, align="right", width=w)
thinkplot.Hist(m_hist, align="left", width=w)
thinkplot.Config(xlabel="Total income", ylabel="Respondent count")
```

## Question 6

```python
def describe(df, typ):
    resp_text = "among {} respondents".format(typ)
    print("standart deviation of total income {} is {}".format(resp_text, df["totincr"].std()))
    print("variance of total income {} is {}".format(resp_text, df["totincr"].var()))
    print("mean total income {} is {}".format(resp_text, df["totincr"].mean()))
    print("minimum total income {} is {}".format(resp_text, df["totincr"].min()))
    print("maximum total income {} is {}".format(resp_text, df["totincr"].max()))
    
describe(married, "married")
print("\n")
describe(neverMarried, "nonmarried")
```
